```python
# Set up the plot
fig, axes = plt.subplots(4,1, figsize=(5, 10))

measurements = ['Tartrate_gL', 'Malate_gL', 'Glucose_gL', 'Fructose_gL']

# Create stripplot for each measurement (swarmplots cannot fit all dots)
for i, measurement in enumerate(measurements):
    sns.stripplot(ax=axes[i], data=hplc_data[hplc_data['Date'] != '2021-08-31'], y=measurement, x='Date', hue='Plot_No', palette='tab20b' , )
    axes[i].set(xlabel=None)
    if i < len(measurements) - 1:
        axes[i].tick_params(bottom=False)
        axes[i].set(xticklabels=[])
    axes[i].get_legend().remove()

# Add a single legend to the side
handles, labels = axes[1].get_legend_handles_labels()
fig.legend(handles, labels, loc='center left', bbox_to_anchor=(1, 0.5), title='Plots')

# Add x-axis labels only to the bottom subplot
for label in axes[-1].get_xticklabels():
    label.set_rotation(90)
    label.set_horizontalalignment('right')

plt.tight_layout()
plt.show()
```

## ANOVA of plots, phenostage and year 

```python
# ANOVA only for veraison samples 
measurement_col = [ 'Brix_degree', 'Tartrate_gL', 'Malate_gL', 'Glucose_gL', 'Fructose_gL']
interaction_terms = ['Plot_No', 'Year', 'Plot_No:Year']
anova_results_df = pd.DataFrame(index=interaction_terms, columns=measurement_col)


for col in measurement_col: 
    # Perform nested ANOVA (specifies both the main effects of individual categories as well as their interaction term)
    model = ols((str(col)+'~ Plot_No * Year'), data=hplc_data[hplc_data['Pheno_Stage']=='Harvest']).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    # Extract p-values and store them in the DataFrame
    for term in interaction_terms:
        if term in anova_table.index:
            anova_results_df.loc[term, col] = anova_table.loc[term, 'PR(>F)']
        else:
            anova_results_df.loc[term, col] = np.nan

print('ANOVA Harvest')
anova_results_df
```

### Check out all plot metadata

```python
# ANOVA 
interaction_terms = ['Year:Plot_No', 'Year:Altitude', 'Year:Average_slope', 'Year:Exposition', 'Year:Average_radiation', 'Year:Soil_thickness', 'Year:Soil_type', 'Year:Soil_depth', 'Year:Hydromorphie', 'Year:Geology']
anova_results_df = pd.DataFrame(index=interaction_terms, columns=measurement_col)


# Loop over each  column and perform ANOVA
for col in measurement_col: 
    # Perform nested ANOVA
    formula = f"{col} ~ Year:Plot_No + Year:Altitude + Year:Average_slope + Year:Exposition + Year:Average_radiation + Year:Soil_thickness + Year:Soil_type + Year:Soil_depth + Year:Hydromorphie + Year:Geology"
    model = ols(formula, data=merged_data_harvest).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    
    # Extract p-values and store them in the DataFrame
    for term in interaction_terms:
        if term in anova_table.index:
            anova_results_df.loc[term, col] = anova_table.loc[term, 'PR(>F)']
        else:
            anova_results_df.loc[term, col] = np.nan

# Convert DataFrame to float
anova_results_df = anova_results_df.astype(float)

# Create a mask for cells with p-values >= 0.05
mask = anova_results_df >= 0.05

# Custom colormap: significant values colored, non-significant values white
cmap = ListedColormap(['white', 'white', 'white'])

# Plot the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(anova_results_df, mask=mask, annot=True, cmap='Blues_r', cbar=False, alpha=0.5,
            linewidths=0.5, linecolor='black', fmt='.3f', 
            annot_kws={"color": "black"})

# Highlight significant p-values with a different colormap
sns.heatmap(anova_results_df, mask=~mask, annot=True, cmap=cmap, cbar=False, 
            linewidths=0.5, linecolor='black', fmt='.3f', 
            annot_kws={"color": "black"})

# Set title and labels
plt.title('ANOVA p-values of Harvest samples (nested for Year)')
plt.xlabel(' ')
plt.ylabel(' ')

# Show the plot
plt.show()
```

### Swarmplot

```python
# Get the column names
# CitrF_gL is omitted because expect for one all measurements are n.d. 
must_measurements = ['TAV_percVol', 'GluFru_gL', 'DegTotal_percVol', 'AT_gL', 'pH',
       'AcMal_gL', 'AcTart_gL', 'KF_mgL',
       'NOPAF_mgL', 'NH3F_mgL', 'Nass_mgL', 'D13']

#Calculate the number of rows and columns for subplots
num_cols = len(must_measurements)
num_rows = (num_cols + 3) // 4  

# Create a figure and axes for subplots
fig, axes = plt.subplots(num_rows, 4, figsize=(12, 3*num_rows))  

# Flatten the axes array if there's only one row
if num_rows == 1:
    axes = axes.reshape(1, -1)

# Loop over each column
for i, column in enumerate(must_measurements):
    # Calculate the row and column index for the subplot
    row_index = i // 4
    col_index = i % 4

    # Create a scatterplot for the current column in the corresponding subplot
    sns.swarmplot(data=must2023_data, y=column, hue='Plot', 
                 ax=axes[row_index, col_index], palette='tab20b') 
    axes[row_index, col_index].set_title(f"{column}")
    axes[row_index, col_index].set_ylabel(column)

plt.tight_layout()
plt.show()
```
