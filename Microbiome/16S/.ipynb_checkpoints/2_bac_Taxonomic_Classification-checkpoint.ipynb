# Filter out non-target seqs


```python
taxonomy = q2.Artifact.load('taxonomy.qza')
decontam_table = q2.Artifact.load('decontam-table.qza')
md = q2.Metadata.load('/home/lfloerl/microterroir/Microbiome/Metadata/16S_md.tsv')
```

```python
exclude_terms = 'mitochondria,chloroplast,Eukaryota,Archaea'
#exclude_terms = 'mitochondria,chloroplast,Alcaligenaceae,Caulobacteraceae,Comamonadaceae,Eukaryota,Pedobacter,Escherichia-Shigella,Archaea'
```

```python
taxa_filtered_table, = q2t.actions.filter_table(
                                table=decontam_table, 
                                taxonomy=taxonomy,
                                exclude=exclude_terms)

phylum_filtered_table, = q2t.actions.filter_table(
                                table=taxa_filtered_table,
                                taxonomy=taxonomy,
                                include='p__')

feature_filtered_table, = filter_samples(
                                table=phylum_filtered_table,
                                min_features=10)

# filter the table to only contain samples listed in the metadata (remove ctrls and PNA test)
md_filtered_table, = filter_samples(table=feature_filtered_table,
                                metadata=md)

taxa_filtered_table.save('chloroplast_mitochondria_filtered_table.qza')
phylum_filtered_table.save('phylum_filtered_table.qza')
feature_filtered_table.save('feature_filtered_table.qza')
md_filtered_table.save('taxa_filtered_table.qza')
```

## Get info on how many reads we lost per sample from filtering 

```python
# export the denoised reads 
!qiime tools export --input-path bac-dada2-2/dada-stats.qzv --output-path bac-dada2-2/dada-stats
```

```python
all_reads = pd.read_csv('bac-dada2-2/dada-stats/metadata.tsv', sep='\t', skiprows=[1])
all_reads.head()
```

```python
# export the number of filtered reads
!qiime tools export --input-path taxa_filtered_table.qzv --output-path taxa_filtered_table
```

```python
# Load the HTML file
with open('taxa_filtered_table/sample-frequency-detail.html', 'r') as file:
    soup = BeautifulSoup(file, 'lxml')

# check out the HTML
#print(soup.prettify()) 

# Find the <script> tag with type="application/json" and id="table-data"
script_tag = soup.find('script', type='application/json', id='table-data')

# Check if the script tag was found
if script_tag:
    # Extract the JSON content from the <script> tag
    json_data = json.loads(script_tag.string)
    
    # Convert the JSON data to a Pandas DataFrame
    filtered_reads = pd.DataFrame(list(json_data.items()), columns=['Sample_ID', 'Frequency'])
    
else:
    print("No script tag with type 'application/json' found.")

# Convert 'Frequency' to numeric
filtered_reads['Frequency'] = filtered_reads['Frequency'].str.replace(',', '').astype(int)

# Define a mapping from sample-id to Frequency in filtered_reads
frequency_map = filtered_reads.set_index('Sample_ID')['Frequency'].to_dict()
```

```python
# Create a new DataFrame to hold the results
results = pd.DataFrame()

# Merge data based on Sample_ID
# Here we assume 'sample-id' in all_reads corresponds to 'Sample_ID' in filtered_reads
for index, row in all_reads.iterrows():
    sample_id = row['sample-id']
    non_chimeric_count = row['non-chimeric']
    
    # Extract the matching Frequency from filtered_reads
    filtered_count = frequency_map.get(sample_id, None)
    
    if filtered_count is not None:
        # Calculate the percentage of reads lost
        percentage_lost = ((non_chimeric_count - filtered_count) / non_chimeric_count) * 100
        
        # Append results to the DataFrame
        results = results.append({
            'Sample_ID': sample_id,
            'Percentage_Reads_Lost': percentage_lost
        }, ignore_index=True)
```

```python
results.set_index('Sample_ID', inplace=True)
results
```

```python
# this is the metadata we use 
md_bac = pd.read_csv('/home/lfloerl/RxCS/Metadata/16S_metadata_filtered_nonFermented.tsv', sep='\t', index_col=0)

# join with the percentage calc. 
md_bac = md_bac.join(results) 

md_bac.to_csv('/home/lfloerl/RxCS/Metadata/16S_metadata_filtered_nonFermented_wLostReads.tsv', sep='\t')
```
