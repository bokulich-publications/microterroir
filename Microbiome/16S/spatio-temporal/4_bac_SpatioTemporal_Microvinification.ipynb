```python
%%bash

# Define the list of sample types and years
sample_types=('Post-MLF' 'Post-AF' 'MV-must')
years=('Year2021' 'Year2022' 'Year2023')

# Path to the metadata file
metadata_file="/home/lfloerl/microterroir/Microbiome/Metadata/16S_lavaux.tsv"

# Create a temporary file to store combinations that exist in the metadata
valid_combinations="valid_combinations.txt"

# Clear the file if it already exists
> $valid_combinations

# Check metadata for available combinations
for samples in "${sample_types[@]}"; do
  for year in "${years[@]}"; do
    # Check if the combination exists in the metadata
    match_count=$(awk -F'\t' -v sample_type="$samples" -v year_cat="$year" \
      '$0 ~ sample_type && $0 ~ year_cat {count++} END {print count}' "$metadata_file")
    
    # If the combination is found, log it
    if [[ $match_count -gt 0 ]]; then
      echo "$samples,$year" >> $valid_combinations
    fi
  done
done

# Read valid combinations and process only these
while IFS=, read -r samples year; do
  echo "Processing $samples in $year"
  
  # Create directory for the sample type and year and navigate into it
  dir_name="${samples}_${year}"
  mkdir -p "./$dir_name"
  pushd "./$dir_name" > /dev/null

  # Filter table and representative sequences
  qiime feature-table filter-samples \
    --i-table /home/lfloerl/cloud/lfloerl/Microterroir/artifacts/16S/lavaux/mv_filtered_table.qza \
    --m-metadata-file "$metadata_file" \
    --p-where "[sample_type]='$samples' AND [Year_Cat]='$year'" \
    --o-filtered-table "mv_${samples}_${year}_filtered_table.qza"

  qiime feature-table filter-seqs \
    --i-data /home/lfloerl/cloud/lfloerl/Microterroir/artifacts/16S/bac-dada2-single/dada-rep-seqs-220-ee4-fa4.qza \
    --i-table "mv_${samples}_${year}_filtered_table.qza" \
    --o-filtered-data "mv_${samples}_${year}_filtered_rep_seqs.qza"

  # Core diversity metrics
  qiime diversity core-metrics \
    --i-table "mv_${samples}_${year}_filtered_table.qza" \
    --p-sampling-depth 150 \
    --m-metadata-file "$metadata_file" \
    --p-n-jobs 10 \
    --p-ignore-missing-samples \
    --output-dir cm150/

  # Kmer diversity metrics
  qiime kmerizer core-metrics \
    --i-sequences "mv_${samples}_${year}_filtered_rep_seqs.qza" \
    --i-table "mv_${samples}_${year}_filtered_table.qza" \
    --p-sampling-depth 150 \
    --m-metadata-file "$metadata_file" \
    --p-n-jobs auto \
    --output-dir kmer_cm150/

  # Export all distance matrices for Mantel
  qiime tools export --input-path cm150/bray_curtis_distance_matrix.qza --output-path cm150/bray_curtis_distance_matrix
  qiime tools export --input-path cm150/jaccard_distance_matrix.qza --output-path cm150/jaccard_distance_matrix
  qiime tools export --input-path kmer_cm150/bray_curtis_distance_matrix.qza --output-path kmer_cm150/bray_curtis_distance_matrix
  qiime tools export --input-path kmer_cm150/jaccard_distance_matrix.qza --output-path kmer_cm150/jaccard_distance_matrix

  # Return to the parent directory
  popd > /dev/null
done < $valid_combinations

# Clean up temporary file
rm $valid_combinations
```

###  Mantel tests 

```python
import sys
sys.path.append('/home/lfloerl/microterroir/Microbiome/Other_scripts/')
from mantel_test import mantel_test_and_plot

!qiime tools export --input-path /home/lfloerl/cloud/lfloerl/Microterroir/artifacts/16S/geodesic_distance_16S.qza --output-path geodesic_distance_16S
```

```python
# Geodesic distance matrix path
geodesic_path = 'geodesic_distance_16S/distance-matrix.tsv'

# Path to the metadata file
metadata_file = '/home/lfloerl/microterroir/Microbiome/Metadata/16S_lavaux.tsv'

# Function to get valid combinations of sample types and years from the metadata
def get_valid_combinations(metadata_file, sample_types, years):
    valid_combinations = []
    metadata = pd.read_csv(metadata_file, sep='\t')  # Adjust delimiter if necessary

    for sample_type in sample_types:
        for year in years:
            matches = metadata[(metadata['sample_type'] == sample_type) & (metadata['Year_Cat'] == year)]
            if not matches.empty:
                valid_combinations.append((sample_type, year))

    return valid_combinations

# Function to run Mantel tests for valid combinations
def run_mantel_tests(valid_combinations, geodesic_path):
    all_results = []

    for sample_type, year in valid_combinations:
        print(f"Processing {sample_type} for {year}...")
        
        # Directory structure for sample type and year
        dir_prefix = f"{sample_type}_{year}/cm150"
        kmer_dir_prefix = f"{sample_type}_{year}/kmer_cm150"
        
        # Define paths for distance matrices
        bray_curtis_path = os.path.join(dir_prefix, 'bray_curtis_distance_matrix', 'distance-matrix.tsv')
        jaccard_path = os.path.join(dir_prefix, 'jaccard_distance_matrix', 'distance-matrix.tsv')
        bray_curtis_kmer_path = os.path.join(kmer_dir_prefix, 'bray_curtis_distance_matrix', 'distance-matrix.tsv')
        jaccard_kmer_path = os.path.join(kmer_dir_prefix, 'jaccard_distance_matrix', 'distance-matrix.tsv')

        # Run Mantel tests for Bray-Curtis and Jaccard matrices
        rho_bray, p_bray = mantel_test_and_plot(
            dm_geodesic_path=geodesic_path, dm_other_path=bray_curtis_path,
            x_label='Geodesic Distance', y_label='Bray-Curtis Distance')
        plt.show()

        rho_jaccard, p_jaccard = mantel_test_and_plot(
            dm_geodesic_path=geodesic_path, dm_other_path=jaccard_path,
            x_label='Geodesic Distance', y_label='Jaccard Distance')
        plt.show()

        rho_bray_kmer, p_bray_kmer = mantel_test_and_plot(
            dm_geodesic_path=geodesic_path, dm_other_path=bray_curtis_kmer_path,
            x_label='Geodesic Distance', y_label='Kmer Bray-Curtis Distance')
        plt.show()

        rho_jaccard_kmer, p_jaccard_kmer = mantel_test_and_plot(
            dm_geodesic_path=geodesic_path, dm_other_path=jaccard_kmer_path,
            x_label='Geodesic Distance', y_label='Kmer Jaccard Distance')
        plt.show()

        # Store the results in a dictionary
        results = {
            'Sample_Type': sample_type,
            'Year': year,
            'Test': ['Bray-Curtis', 'Jaccard', 'Kmer Bray-Curtis', 'Kmer Jaccard'],
            'Rho': [rho_bray, rho_jaccard, rho_bray_kmer, rho_jaccard_kmer],
            'p-val': [p_bray, p_jaccard, p_bray_kmer, p_jaccard_kmer]
        }

        # Convert to dataframe and add to list of results
        results_df = pd.DataFrame(results)
        all_results.append(results_df)

    # Concatenate all results into one dataframe
    final_results_df = pd.concat(all_results, ignore_index=True)
    return final_results_df

# Define sample types and years
sample_types = ['Post-MLF', 'Post-AF', 'MV-must']
years = ['Year2021', 'Year2022', 'Year2023']

# Get valid combinations from the metadata
valid_combinations = get_valid_combinations(metadata_file, sample_types, years)

# Run the Mantel tests for valid combinations
final_results_df = run_mantel_tests(valid_combinations, geodesic_path)
```
