```python
md = pd.read_csv('ITS_lavaux.tsv', sep='\t')
md_valais = pd.read_csv('ITS_Valais_and_Lavaux23.tsv', sep='\t')

md_valais['Plot_ID'] = md_valais['Plot']
md_valais['Year'] = '2023'
md_valais['sample_type'] = 'must'

md['Region'] = 'Lavaux'


md = md[['id', 'sample_type', 'Year', 'Plot_ID', 'Region']]
md_valais = md_valais[['id', 'sample_type', 'Year', 'Plot_ID', 'Region']]

md_merged = pd.concat([md, md_valais])
md_merged = md_merged.drop_duplicates(subset='id')

md_merged.head()
```

```python
md_merged.shape
```

```python
md_merged['Region'].value_counts()
```

```python
# Table 1: Samples per sample type
samples_per_type = md_merged['sample_type'].value_counts().reset_index()
samples_per_type.columns = ['sample_type', 'sample_count']

# Table 2: Samples per year
samples_per_year = md_merged['Year'].value_counts().reset_index()
samples_per_year.columns = ['Year', 'sample_count']

# Table 3: Samples per plot
samples_per_plot = md_merged['Plot_ID'].value_counts().reset_index()
samples_per_plot.columns = ['Plot_ID', 'sample_count']

# Print all three tables
print("Samples per sample type:")
print(samples_per_type, end='\n\n')

print("Samples per year:")
print(samples_per_year, end='\n\n')

print("Samples per plot:")
print(samples_per_plot)
```

```python
# Create the pivot table
summary_table = pd.pivot_table(md_merged, 
                                index="sample_type", 
                                columns="Year", 
                                values="id", 
                                aggfunc="count", 
                                fill_value=0)

# Reset index if needed
summary_table = summary_table.reset_index()

# Display the result
summary_table
```

```python
# Filter for Lavaux in 2021
filtered_df = md_merged[(md_merged['Region'] == 'Lavaux') & (md_merged['Year'] == 2021)]

# Count samples per plot
samples_per_plot = filtered_df['Plot_ID'].value_counts()

# Get min and max
min_samples = samples_per_plot.min()
max_samples = samples_per_plot.max()

print(f"Minimum samples per plot (Lavaux 2021): {min_samples}")
print(f"Maximum samples per plot (Lavaux 2021): {max_samples}")
```
