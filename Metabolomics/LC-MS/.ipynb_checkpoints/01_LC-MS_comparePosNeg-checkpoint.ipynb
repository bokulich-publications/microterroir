# Compare LC-MS positive and negative mode measurements


```python
import pandas as pd
import numpy as np
import os

# Visualizations 
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
from sklearn.decomposition import PCA


%matplotlib inline
```

```python
# set the directory 
os.chdir('/home/lfloerl/cloud/lfloerl/Microterroir/LC-MS_data/Results')
```

```python
sample_type_palette = {'Must': '#482677FF', 'PostAF': '#238A8DFF','PostMLF': '#74D055FF'}
```

# All Features

```python
df_neg_all = pd.read_csv('Neg_All_OutlierRemoved.csv')
df_pos_all = pd.read_csv('Pos_All_OutlierRemoved.csv')
```

```python
print("All positive mode", df_pos_all.shape)
print("All negative mode", df_neg_all.shape)
```

```python
# Get the column names excluding the first 4 columns
cols_neg = df_neg_all.columns[4:]  # Excluding first 4 columns
cols_pos = df_pos_all.columns[4:] 
# Convert columns to sets
IDs_neg = set(cols_neg)
IDs_pos = set(cols_pos)

# Create the Venn diagram
venn2(subsets=[IDs_pos, IDs_neg], set_labels=('Positive Mode', 'Negative Mode'),
      set_colors=('#006994', '#7FB3D5'), alpha=0.5)

# Display the plot
plt.title('Venn Diagram of All Features')
plt.show()
```

## Neg

```python
# make PCA df 
pca_df = df_neg_all.iloc[:, 4:]

# Define the number of components to test
n_components = 10

pca = PCA(n_components=n_components)
pca_scores = pca.fit_transform(pca_df)

# Extract explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Create subplots for explained variance and cumulative variance plots
fig, axes = plt.subplots(1, 2, figsize=(10,2.5))

# Bar plot for variance explained per component
axes[0].bar(range(1, n_components + 1), explained_variance_ratio, color='#20A386FF', alpha=0.8)
axes[0].set_xlabel('Principal Component', fontsize=11)
axes[0].set_ylabel('Var. Explained', fontsize=11)
axes[0].set_xticks(range(1, n_components + 1))
axes[0].grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)
# Add numeric values on top of bars
for p in axes[0].patches:
    axes[0].annotate(f'{p.get_height():.2f}', 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=10, fontweight='bold')

# Line plot for cumulative explained variance
axes[1].plot(range(1, n_components + 1), np.cumsum(explained_variance_ratio), marker='o', linestyle='-', color='#482677FF')
axes[1].set_xlabel('Principal Component', fontsize=11)
axes[1].set_ylabel('Cum. Var. Explained', fontsize=11)
axes[1].set_xticks(range(1, n_components + 1))
axes[1].grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Adjust layout to ensure proper display
plt.tight_layout()
plt.show()
```

```python
# Perform PCA
n_components = 2
pca = PCA(n_components=n_components)
pca_scores = pca.fit_transform(pca_df)

# Create a new DataFrame for PCA results
pca_results = pd.DataFrame(pca_scores, columns=["PC1", "PC2"], index=pca_df.index)

# Merge 'year' and 'sample_type' columns back to PCA results
pca_results[["year", "sample_type"]] = df_neg_all.loc[:, ["year", "sample_type"]].values  

# Get explained variance
explained_var_pc1 = pca.explained_variance_ratio_[0] * 100  # Convert to percentage
explained_var_pc2 = pca.explained_variance_ratio_[1] * 100  # Convert to percentage

# Plot PCA results
plt.figure(figsize=(2.5,2.5))
sns.scatterplot(
    data=pca_results, 
    x="PC1", 
    y="PC2", 
    hue="sample_type",
    style="year",
    palette=sample_type_palette, 
    alpha=0.8)

plt.xlabel(f"PC1 ({explained_var_pc1:.2f}%)")
plt.ylabel(f"PC2 ({explained_var_pc2:.2f}%)")

# Adjust legend
plt.legend(title="Sample Type & Year", bbox_to_anchor=(1.05, 1), loc='upper left')  
plt.grid(True, linestyle="--", alpha=0.5)

plt.show()
```

#### Merge but keep the higher value from the overlapping col

```python
Neg_MS2 = pd.read_csv('Neg_MS2_OutlierRemoved.csv')
Pos_MS2 = pd.read_csv('Pos_MS2_OutlierRemoved.csv')

Pos_MS2.set_index('Sample Name', inplace=True)
Pos_MS2.drop(columns=['sample_type', 'year', 'plot'], inplace=True)
Neg_MS2.set_index('Sample Name', inplace=True)
Neg_MS2.drop(columns=['sample_type', 'year', 'plot'], inplace=True)
```

```python
# Step 1: Find matching Sample Names
matching_samples = set(Pos_MS2.index) & set(Neg_MS2.index)

# Filter both dataframes to keep only matching rows
Pos_MS2_matched = Pos_MS2[Pos_MS2.index.isin(matching_samples)]
Neg_MS2_matched = Neg_MS2[Neg_MS2.index.isin(matching_samples)]

# Step 2: Identify common and unique columns
common_cols = set(Pos_MS2_matched.columns) & set(Neg_MS2_matched.columns)
unique_pos_cols = set(Pos_MS2_matched.columns) - common_cols
unique_neg_cols = set(Neg_MS2_matched.columns) - common_cols

# Step 4: Determine which dataframe has the higher total count for each overlapping column
higher_total_cols = {}
for col in common_cols:
    pos_sum = Pos_MS2_matched[col].fillna(0).sum().sum()  # Convert to scalar
    neg_sum = Neg_MS2_matched[col].fillna(0).sum().sum()  # Convert to scalar
    higher_total_cols[col] = 'Pos_MS2' if pos_sum > neg_sum else 'Neg_MS2'
    
# Step 5: Select columns from the dataframe with the higher total
Pos_selected = Pos_MS2_matched[list(unique_pos_cols) + [col for col, source in higher_total_cols.items() if source == 'Pos_MS2']]
Neg_selected = Neg_MS2_matched[list(unique_neg_cols) + [col for col, source in higher_total_cols.items() if source == 'Neg_MS2']]

# Step 5: Merge all selected columns
merged_df = Pos_selected.merge(Neg_selected, left_index=True, right_index=True, how='outer')

merged_df.head()
```

```python
merged_df.to_csv('PosNeg_MS2_merge_noOverlap.tsv', sep='\t')
```
