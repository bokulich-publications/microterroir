# Compare LC-MS positive and negative mode measurements


```python
import pandas as pd
import numpy as np
import os

# Visualizations 
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

%matplotlib inline
```

```python
# set the directory 
os.chdir('/home/lfloerl/cloud/lfloerl/Microterroir/LC-MS_data/Results')
```

```python
df_neg_all = pd.read_csv('Neg_All_OutlierRemoved.csv')
df_pos_all = pd.read_csv('Pos_All_OutlierRemoved.csv')
```

```python
print("All positive mode", df_pos_all.shape)
print("All negative mode", df_neg_all.shape)
```

```python
# Get the column names excluding the first 4 columns
cols_neg = df_neg_all.columns[4:]  # Excluding first 4 columns
cols_pos = df_pos_all.columns[4:] 
# Convert columns to sets
IDs_neg = set(cols_neg)
IDs_pos = set(cols_pos)

# Create the Venn diagram
venn2(subsets=[IDs_pos, IDs_neg], set_labels=('Positive Mode', 'Negative Mode'),
      set_colors=('#006994', '#7FB3D5'), alpha=0.5)

# Display the plot
plt.title('Venn Diagram of All Features')
plt.show()
```
